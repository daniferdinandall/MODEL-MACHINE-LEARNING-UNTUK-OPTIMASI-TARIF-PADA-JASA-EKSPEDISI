```python
import json
import pandas as pd
import numpy as np

# Path ke file JSON
dataset_kota = '../dataset/dataset_kota.json'
path_dataset = "../dataset/SAMPLE DATA KC AMUNTAI JAN-JUN 2024 data Penelitian.xls"
```

```python
# Membaca file JSON
with open(dataset_kota, 'r') as file:
    data_kordinat = json.load(file)

# Menghitung panjang array
array_length = len(data_kordinat)

# Menampilkan panjang array
print(f"Panjang array: {array_length}")
```

```python
# Mengonversi data JSON ke DataFrame pandas
df_kordinat = pd.DataFrame(data_kordinat)

# Menampilkan DataFrame
df_kordinat.head()
```

```python
# Mengubah semua teks dalam kolom 'Kota Penerima' menjadi huruf kecil
df_kordinat['name'] = df_kordinat['name'].str.lower()
df_kordinat['alt_name'] = df_kordinat['alt_name'].str.lower()
```

```python
# Membaca file .xls dengan baris pertama sebagai header
df = pd.read_excel(path_dataset, header=0)
# Menghapus baris kedua (baris ke-1 dalam indeks 0)
df = df.drop(index=0)
df = df.drop(index=[1, 2])
# Menampilkan DataFrame
df.head()
```

```python

# Menampilkan panjang DataFrame (jumlah baris tidak termasuk header)
print(f"Panjang DataFrame (jumlah baris): {len(df)}")
```

```python
# Mengganti " , " dengan "," pada kolom 'Kota Penerima'
df['Kota Penerima'] = df['Kota Penerima'].str.replace(' , ', ',', regex=False)
df['Kota Penerima'] = df['Kota Penerima'].str.replace(' ,', ',', regex=False)
df['Kota Penerima'] = df['Kota Penerima'].str.replace(', ', ',', regex=False)

# Mengubah semua teks dalam kolom 'Kota Penerima' menjadi huruf kecil
df['Kota Penerima'] = df['Kota Penerima'].str.lower()

# replace "kab" dengan "kabupaten"
df['Kota Penerima'] = df['Kota Penerima'].str.replace('kab', 'kabupaten', regex=False)
df['Kota Penerima'] = df['Kota Penerima'].str.replace('kabupaten.', 'kabupaten', regex=False)
```

```python
df["Kota Penerima"].head()
```

```python
# Fungsi untuk mengambil bagian yang dimulai dengan 'kota' atau 'kabupaten'
def extract_kota_or_kabupaten(text):
    parts = text.split(',')
    for part in parts:
        if part.startswith('kota '):
            return part
        elif part.startswith('kabupaten '):
            return part
    return text

# Terapkan fungsi ke kolom 'Kota Penerima' dan simpan hasilnya di kolom baru 'Kota'
df['Kota'] = df['Kota Penerima'].apply(extract_kota_or_kabupaten)

df.head()
```

```python
# Menggabungkan df dengan df_kordinat berdasarkan kecocokan kolom 'Kota' dengan 'name'
df = df.merge(df_kordinat[['name', 'longitude', 'latitude']], left_on='Kota', right_on='name', how='left')

# Menggabungkan df dengan df_kordinat berdasarkan kecocokan kolom 'Kota' dengan 'alt_name' jika tidak ditemukan di 'name'
df = df.merge(df_kordinat[['alt_name', 'longitude', 'latitude']], left_on='Kota', right_on='alt_name', how='left', suffixes=('', '_alt'))

# Mengisi NaN dengan data dari kolom 'longitude_alt' dan 'latitude_alt' jika kolom 'longitude' dan 'latitude' NaN
df['longitude'] = df['longitude'].fillna(df['longitude_alt'])
df['latitude'] = df['latitude'].fillna(df['latitude_alt'])

# Menghapus kolom yang tidak diperlukan
df = df.drop(columns=['name', 'alt_name', 'longitude_alt', 'latitude_alt'])

# Menampilkan hasil akhir
# print(df["longitude"].isna().sum())
df2=df[df['longitude'].isnull()]
df2
```

```python
# Cek baris yang memiliki nilai None atau NaN di kolom Kota
print("Baris dengan nilai None atau NaN di kolom Kota:")
print(df["longitude"].isna().sum())
df[df['longitude'].isna()].head()
```

```python
def check_similarity(kota, df_kordinat):
    for _, row_kordinat in df_kordinat.iterrows():
        if row_kordinat['name'] in kota:
            return row_kordinat['longitude'], row_kordinat['latitude']
        elif row_kordinat['alt_name'] in kota:
            return row_kordinat['longitude'], row_kordinat['latitude']
    return np.nan, np.nan

# Loop melalui setiap baris di df dan perbarui kolom longitude dan latitude berdasarkan kesamaan kata
x=0
for i, row in df.iterrows():
    if pd.isna(row['longitude']):
        longitude, latitude = check_similarity(row['Kota'], df_kordinat)
        df.at[i, 'longitude'] = longitude
        df.at[i, 'latitude'] = latitude
df.head()
```
