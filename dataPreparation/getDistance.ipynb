```python
# Importing the geodesic module from the library 
from geopy.distance import geodesic 
import pandas as pd
```

```python
df=pd.read_csv("../dataset/data_kordinat.csv")
```

```python
df["Kantor Asal"].unique()
# jika di cari di google maps, kantor asal berada di latitude: -2.4156183 dan longitude: 115.2397901
```

```python
def getDistance(latitude, longitude):
  coords_penerima = (latitude, longitude)
  coords_pengirim = (-2.4156183,115.2397901)
  return geodesic(coords_penerima, coords_pengirim).km

df['Jarak (KM)'] = df.apply(lambda x: getDistance(x['latitude'], x['longitude']), axis=1)
df.head()
```

```python
df.to_csv("../dataset/data_jarak.csv", index=False)
```
